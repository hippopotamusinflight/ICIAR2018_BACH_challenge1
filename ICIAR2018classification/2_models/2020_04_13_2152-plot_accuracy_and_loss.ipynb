## 2020_04_13_2152-plot_accuracy_and_loss

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
classification_dir = "/global/home/hpc4535/CISC881FinalProject/ICIAR2018classification/"
csv_dir = os.path.join(classification_dir, "2_pretrained_models_test/outputs/")
history_csv = "ResNet50_wDropout_p1400s100_73split_5plus150epochs.csv"
```

```python
colnames = ["epoch","val_loss","val_accuracy","train_loss","train_accuracy"]
history = pd.read_csv(os.path.join(csv_dir, history_csv), skiprows=1, names=colnames)
```

```python
history
```

```python
plt.plot(history["val_loss"])
plt.plot(history["train_loss"])
plt.title("training and validation loss, \nResNet50, p1400s100")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["Validation", "Training"], loc="upper left")
plt.show()
```

```python
plt.plot(history["val_accuracy"])
plt.plot(history["train_accuracy"])
plt.title("training and validation accuracy, \nResNet50, p1400s100")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["Validation", "Training"], loc="upper left")
plt.show()
```
